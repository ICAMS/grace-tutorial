## Overview

This notebook validates the **fine-tuned GRACE-2L foundation model** on Al-Li data.
After fine-tuning a pretrained GRACE model (`GRACE-2L-OMAT`) on target DFT data,
we evaluate its accuracy by computing energy and force parity plots, error tables,
and comparing it against the reference dataset and the base foundation model.

```python
%pylab inline
```

```python
import pandas as pd
```

```python
from amstools import *
```

```python
from tensorpotential.calculator import grace_fm, TPCalculator
```

```python
from amstools.thermodynamics import run_convex_hull_calculation, plot_convex_hull
```

```python
from amstools.sources import fetch_structures, fetch_mp_reference_df
```

```python
from ase.build import bulk
```

```python
from glob import glob
```

## Step 1 — Load Reference Structures & DFT Convex Hull

Fetch Al-Li reference structures from the Materials Project cache. These serve as the
**ground truth** for the convex hull comparison.

```python
cache_dir='../1-AlLi-GRACE-2LAYER/0-data/AlLi_Materials_Project'
```

```python
ref_df=fetch_mp_reference_df(['Al','Li'], 
                             # mp_api_key='XXXXX',  ## get from https://next-gen.materialsproject.org/api                         
                             max_atoms=32,
                             cache_dir=cache_dir)
```

```python
ref_df
```

```python
struct_dict=fetch_structures(['Al','Li'], 
                             # mp_api_key=MP_API_KEY, 
                             max_atoms=32,
                             cache_dir=cache_dir)
```

## Step 2 — Convex Hull: Fine-tuned GRACE-2L-OMAT Model

Run the convex hull pipeline using the **fine-tuned** foundation model (`seed/1/final_model`)
to assess how well it predicts the Al-Li phase stability.

```python
calc_dict={}
```

```python
models = glob("3a-finetuning*/seed/1/final_model")
models
```

```python
for path in models:
    if path not in calc_dict:
        calc_dict[path]=TPCalculator(path)
        print("Loaded",path)
```

```python
dfs_dict={}
```

```python
for path, calc in calc_dict.items():
    if path not in dfs_dict:
        print("="*40)
        print(path)
        df, pipeline_dict=run_convex_hull_calculation(
            structure_dict=struct_dict,
            calc=calc, 
            # pipeline_dict=pipeline_dict,
            verbose=False
        )
        dfs_dict[path]=df
```

## Step 3 — Convex Hull: GRACE-2L Trained from Scratch (Baseline)

Compute the convex hull using the model trained **from scratch** on Al-Li data (Part 1 of the tutorial).
This provides a baseline to measure the benefit of fine-tuning vs. training from scratch.

```python
calc_scr=TPCalculator("../1-AlLi-GRACE-2LAYER/1-fit/seed/1/final_model/")
```

```python
pipeline_dict_scr = {}
```

```python
df_scr, pipeline_dict_scr=run_convex_hull_calculation(
    structure_dict=struct_dict,
    calc=calc_scr,
    pipeline_dict=pipeline_dict_scr,
    verbose=False
)
```

## Step 4 — Convex Hull: GRACE-2L-OMAT Foundation Model (Zero-Shot)

Evaluate the **unmodified** foundation model `GRACE-2L-OMAT-medium-ft-E` (no fine-tuning) to
understand how much the fine-tuning step contributed to accuracy improvements.

```python
calc_fm=grace_fm("GRACE-2L-OMAT-medium-ft-E")
```

```python
pipeline_dict_fm = {}
```

```python
df_fm, pipeline_dict_fm=run_convex_hull_calculation(
    structure_dict=struct_dict, 
    calc=calc_fm,
    pipeline_dict=pipeline_dict_fm,
    verbose=False
)
```

## Step 5 — Compare All Convex Hulls

Plot all three models side-by-side against the DFT Materials Project reference to
see the accuracy improvements from fine-tuning vs. training from scratch.

```python
# subdict={n.split("/")[0]:df for n, df in dfs_dict.items()}

# subdict.update({"GRACE-2L-scr":df_scr, 
# "GRACE-2L-OMAT-M-ft-E":df_fm,
#                })
```

```python
df_naive=dfs_dict['3a-finetuning/seed/1/final_model']
```

```python
cdf=pd.read_pickle("AlLi-DFT-convexhull.pkl.gz")
```

```python
for name, df in dfs_dict.items():
    name = "Naive-FT" if "freeze" not in name else "Freeze-FT"
    plot_convex_hull({
        "FM":df_fm,
        name: df,
        # "Scratch":df_scr, 
        
        # "FT-naive":df_naive,
        "DFT": cdf,
        
    }
    )
    plt.show()
```
