## Overview

This notebook discovers all training runs under the current directory, loads their metrics
(energy MAE, force MAE, epoch curves), and produces a **dashboard plot** comparing multiple
GRACE-2L fit seeds. Use it to select the best-performing model or monitor convergence.

```python
import os
os.environ["CUDA_VISIBLE_DEVICES"]="-1"
```

```python
%pylab inline

plt.rcParams["figure.figsize"] = (7,5)
plt.rcParams["figure.dpi"] =  150
```

```python
import pandas as pd
```

```python
from tensorpotential.utils import *
```

## Step 1 — Initialize the Fit Dictionary

Create an empty dictionary that will accumulate metrics from all discovered training runs.

```python
fit_dict={}
```

## Step 2 — Discover & Register Fit Folders

`discovery_fit_folders(".")` scans the current directory tree for completed GRACE fit runs
and returns their paths. Add the discovered folders to `fit_dict` to track them.

```python
folders=discovery_fit_folders(".")
folders
```

```python
# folders+=['../1-AlLi-GRACE-2LAYER/1-fit/seed/1']
```

## Step 3 — Load Metrics & Generate Dashboard

`update_fit_metrics` loads training logs for all registered runs. Re-run this cell
whenever a fit produces new checkpoints to refresh the plots below.

```python
update_fit_metrics(fit_dict, folders)
```

```python
fkey="mae/depa"
```

```python
common_prefix,df, gdf=process_fit_dict(fit_dict, fkey=fkey, align_folder=True)
```

```python
df[["name","mae/depa","mae/f_comp",
   
    "epoch"]].sort_values(fkey).head(50)
```

```python
# gdf
```

```python
fig, ax_e, ax_f = plot_dashboard(fit_dict,
                                 # plot_train=True,
                                 label="shortname_and_seed", 
                                 metric="mae", # 
                                 exclude_list=['./3b-distillation-frozen-ft/seed/1']
                                )
ax_e.legend(ncol=2, bbox_to_anchor=(0.5,-0.15), loc="upper center")
ax_f.legend(ncol=2, bbox_to_anchor=(0.5,-0.15), loc="upper center")

# ax_e.set_ylim(0, 0.06)
# ax_f.set_ylim(0, 0.005)
ax_e.set_xscale('log')
fig.suptitle(common_prefix)
fig.tight_layout()
```
